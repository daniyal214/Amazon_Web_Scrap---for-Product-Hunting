```python
card_bold_items = {'ASIN':[], 'Brand':[], 'Daily_Sales':[], 'Monthly_Revenue':[],
                  'Date_First_Available':[], 'Net':[], 'Fees':[]}
for index, value in enumerate(card_bold):
    if (index%7 == 0) & (sum(map(len, value)) == 10):
        card_bold_items['ASIN'].append(value)
    elif index%7 == 1:
        card_bold_items['Brand'].append(value)
    elif index%7 == 2:
        card_bold_items['Daily_Sales'].append(value)
    elif index%7 == 3:
        card_bold_items['Monthly_Revenue'].append(value)
    elif index%7 == 4:
        card_bold_items['Date_First_Available'].append(value)
    elif index%7 == 5:
        card_bold_items['Net'].append(value)
    else:
        card_bold_items['Fees'].append(value)
```

```python
card_all = driver.find_elements_by_css_selector('.CardWrapper-sc-1bf5o4y-0.dVGGBd')
```

```python
card_bold_items = {'ASIN':[], 'Brand':[], 'Daily_Sales':[], 'Monthly_Revenue':[], 'Monthly_Sales':[],
                  'Date_First_Available':[], 'Net':[], 'Fees':[]}

for i in card_all:
    indv = i.text.replace('\n', ' ').split(':')
    for indx, j in enumerate(indv):
        if indx == 1:
            card_bold_items['ASIN'].append(j)
        elif indx == 2:
            card_bold_items['Monthly_Sales'].append(j)
        elif indx == 3:
            card_bold_items['Brand'].append(j)
        elif indx == 4:
            card_bold_items['Daily_Sales'].append(j)
        elif indx == 5:
            card_bold_items['Monthly_Revenue'].append(j)
        elif indx == 6:
            card_bold_items['Date_First_Available'].append(j)
        elif indx == 7:
            card_bold_items['Net'].append(j)
        elif indx == 8:
            card_bold_items['Fees'].append(j)
```

```python
card_bold_items
```

```python
final = {'ASIN':[], 'Brand':[], 'Daily_Sales':[], 'Monthly_Revenue':[], 'Monthly_Sales':[],
                  'Date_First_Available':[], 'Net':[], 'Fees':[]}
for i in card_bold_items['ASIN']:
    final['ASIN'].append(i.split()[0])
```

```python
len(card_bold_items['Fees'])
```

```python
l = ['Brand', 'Daily_Sales', 'Monthly_Sales', 'Date_First_Available', 'Net']
for j in l:
    for i in card_bold_items[j]:
        final[j].append(i.split()[0])
        
final['Fees'] = card_bold_items['Fees']
```

```python
for i in card_bold_items['Monthly_Revenue']:
    final['Monthly_Revenue'].append(i.split()[0] + i.split()[1])
```
