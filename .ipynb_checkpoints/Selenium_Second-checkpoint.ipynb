```python
titles_lst = []
price_lst = []
review_lst = []
url_lst = []
mo_sale_lst = []
sellers_lst = []
ti = []
ti_pr = []
cnt = {}
count = 1




for i in range(30):
    
    #Product_Title
    time.sleep(3)
    titles = driver.find_elements_by_css_selector('.a-size-mini.a-spacing-none.a-color-base.s-line-clamp-4')
    for title in titles:
        titles_lst.append(title.text)
#     titles_lst = [x for x in titles_lst if x]
    
    
    #Product_Price
    time.sleep(3)
    pr = driver.find_elements_by_xpath('//span[@class = "a-price"]')
    for price in pr:
        price_lst.append(price.text.replace('\n', '.'))
    price_lst = [x for x in price_lst if x]
    
    
    prr = []
    parent_div = driver.find_elements_by_xpath("//div[@class = 'a-section a-spacing-none']")
    for i in parent_div:
        prr.append(len(i.find_elements_by_xpath("./div")))
        prr.append(i.text)
    prr = [x for x in prr if x] 
    found_pr = []
    for index, value in enumerate(prr):
        if (value == 2):
            found_pr.append(prr[index+1])
    for i in found_pr:
        ti_pr.append(i.split('\n')[0])
        
        
    #Seller
    time.sleep(3)
    sellers = driver.find_elements_by_css_selector('.SellerType-sc-1bf5o4y-4.lesqVv')
    for seller in sellers:
        sellers_lst.append(seller.text)
    sellers_lst = [x for x in sellers_lst if x]
        
    
    #Product_Rating
    time.sleep(3)
    reviews = driver.find_elements_by_css_selector('.a-row.a-size-small')
    r = []
    for review in reviews:
        r.append(review.text)
        review_lst.append(review.text)
    review_lst = [x for x in review_lst if x]
    cnt[str(count)] = r

    rev = []
    parent_div = driver.find_elements_by_xpath("//div[@class = 'a-section a-spacing-none']")
    for i in parent_div:
        rev.append(len(i.find_elements_by_xpath("./div")))
        rev.append(i.text)
    rev = [x for x in rev if x]    
    found = []
    for index, value in enumerate(rev):
        if (value == 3):
            found.append(rev[index+1])     
    for i in found:
        if type(i) == str:
            ti.append(i.split('\n')[0])
            
        
    count = count+1

    
    #Monthly_Sale
    m_sales = driver.find_elements_by_xpath('//div[@content = "30-day unit sales estimate"]')
    for m_sale in m_sales:
        mo_sale_lst.append(m_sale.text)
    
    
    try:
        time.sleep(3)
        element = driver.find_element_by_class_name('a-last')
        actions = ActionChains(driver)
        actions.move_to_element(element).perform()
        time.sleep(5)
        element.click()
    except NoSuchElementException:
        break
```
