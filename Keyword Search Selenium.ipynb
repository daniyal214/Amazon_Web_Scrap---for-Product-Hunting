```python
final = {}
for keyword in A[108:]:
    time.sleep(3)
    elem = driver.find_element_by_xpath('//*[@id="twotabsearchtextbox"]')
    elem.send_keys(keyword)
    elem.send_keys(Keys.ENTER) 
    
    screenWidth, screenHeight = pyautogui.size()
    currentMouseX, currentMouseY = pyautogui.position()
    pyautogui.FAILSAFE = False
    time.sleep(5)
    pyautogui.moveTo(1800, 100)
    pyautogui.click()
    time.sleep(3)
    pyautogui.moveTo(1500, 300)
    pyautogui.click()
    time.sleep(10)
    
    
    final[keyword] = {}
    
    
    #MONTHLY REVENIEW
    time.sleep(5)
    mo = driver.find_elements_by_xpath('//div[@class = "rt-tr-group"]//div[8]')
    mo_lst = []
    for i in mo:
        rep = (i.text.replace('£',''))
        mo_lst.append(rep.replace(',',''))

    mo_lst = mo_lst[:10]
    mo_lst = [x if x !='' else '0' for x in mo_lst]

    new_mo = []
    for i in mo_lst:
        new_mo.append(int(i.replace('...', '0')))

    if (sum(True for x in new_mo if x<2000)) > 2:
        final[keyword]['Monthly Revenue'] = 'Rejected'
    else:
        final[keyword]['Monthly Revenue'] = 'Accepted'



    #REVIEWS
    if (final[keyword]['Monthly Revenue']) == 'Accepted':
        divs = driver.find_elements_by_xpath('//div[@class = "rt-tr-group"]//div[11]')

        reviews = []
        time.sleep(3)
        for i in divs:
            reviews.append(i.text)

        reviews = reviews[:10]
        reviews = [x if x != '' else '0' for x in reviews]


        new_reviews = []
        for i in reviews:
            new_reviews.append(int(i.replace(',','')))


        bins = {'Small':0, 'Mid':0, 'High':0}

        for i in new_reviews:
            if i < 1000:
                bins['Small'] +=1
            elif ((i >=1000) & (i<=1200)):
                bins['Mid'] +=1
            else:
                bins['High'] +=1

        review_count = 0
        if ((bins['Mid'] > 3) or (bins['High'] > 0)):
            review_count =  review_count+1
        else:
            review_count = 0

        if review_count != 0:
            final[keyword]['Reviews'] = 'Rejected'
        else:
            final[keyword]['Reviews'] = 'Accepted'

    
    
    
        #BRAND DOMINANCE
        if (final[keyword]['Reviews']) == 'Accepted':
            brnds = driver.find_elements_by_css_selector('.TruncatableLabel-sc-17y9w0e-3.deAMfA')
            brnds = driver.find_elements_by_css_selector('.TruncatableLabel-sc-17y9w0e-3.deAMfA')
            br = []
            for i in brnds:
                    br.append(i.text)

            br = [x if x != '' else 'Unknown' for x in br]
            br = br[:10]

            lib = {}
            for i in br:
                if i not in lib:
                    lib[i] = 1
                else:
                    lib[i] = lib[i]+1
                    
            
            arg = max(lib, key = lib.get)
            if arg != []:
                result = lib[arg] > 3

            if result == True:
                final[keyword]['Brand'] = 'Rejected'
            else:
                final[keyword]['Brand'] = 'Accepted'
    


            #RATINGS / STARS
            if (final[keyword]['Brand']) == 'Accepted':
                stars = driver.find_elements_by_xpath('//div[@class = "rt-tr-group"]//div[12]')
                rats = []
                for i in stars:
                    rats.append(i.text)

                rats = rats[:10]
                rats = [x if x != '' else '4.444' for x in rats]
                rats = [float(x) for x in rats]

                count = 0
                for i in rats:
                    if i<4.5:
                        count+=1

                if count == 0:
                    final[keyword]['Ratings'] = 'Accepted'
                elif count == 1:
                    final[keyword]['Ratings'] = 'Only 1 less than 4.5'
                elif count == 2:
                    final[keyword]['Ratings'] = '2 less than 4.5'
                else:
                    final[keyword]['Ratings'] = 'Rejected'


                #SELLER TYPE
                if (final[keyword]['Ratings']) != 'Rejected':
                    seller = driver.find_elements_by_xpath('//div[@class = "rt-tr-group"]//div[15]')
                    sellers = []
                    for i in seller:
                        sellers.append(i.text)
                    sellers = sellers[:10]
                    sellers = [x if x != '' else 'Unknown' for x in sellers]


                    count=0
                    for i in sellers:
                        if i=='FBA':
                            count+=1
                    if count>=7:
                        final[keyword]['Seller Type'] = 'Accepted'
                    else:
                        final[keyword]['Seller Type'] = 'Rejected'
        
    
    close = driver.find_elements_by_css_selector('.ButtonWrapper-f6l1p1-5.iZomMx')
    close[2].click()
    elem = driver.find_element_by_xpath('//*[@id="twotabsearchtextbox"]')
    elem.clear()
```
